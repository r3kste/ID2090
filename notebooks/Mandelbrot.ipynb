# Mandelbrot

```python
def m(c):
    z = complex(0,0)
    c = complex(c)
    for j in range(100):
        if abs(z) > 2 :
            break
        z = z^2 + c
    else:
            return complex(0,0)
    return complex(0, j)
```

```python
complex_plot(m, (-2,1), (-1.5, 1.5), plot_points=1000)
```

```python
v = vector([1, 0])
var('j')
k = 20
disp = sum(circle(matrix([[cos(j*2*pi/k),-sin(j*2*pi/k)], [sin(j*2*pi/k), cos(j*2*pi/k)]])*v, 1)
                 for j in range(k))
```
