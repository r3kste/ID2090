```python
%matplotlib inline
import math
x=list(range(0,20,1))
```

```python
for i in range(0,20,1):
    x[i] = i*math.pi/10
```

```python
y=list(map(math.sin, x))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x,y,'co-',label='sin')
plt.show()
```
