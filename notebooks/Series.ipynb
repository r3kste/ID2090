# Series

```python
var('x')
s(x) = sin(x)*cos(x)/(x^2)
s.show()
```

```python
p1=plot(s(x),x,[1,5])
p1.show()
```

```python
s1(x) = s.series(x==1, 4)
s1.show()
```

```python
s2(x) = s.series(x==2, 4)
s2.show()
```

```python
p2=plot(s1(x),x,[1,1.5],color='red')
p3=plot(s2(x),x,[1.8,2.2],color='green')
(p1+p2+p3).show()
```

## Taylor series

```python
s3(x) = s.taylor(x,3.5,5)
s3.show()
```

```python
p1a=plot(s(x),x,[1,6])
p4 = plot(s3(x),x,[1,6], color='red')
(p1a+p4).show()
```

```python
f = piecewise([((0,pi/2),-1),((pi/2,pi),2)])
f.fourier_series_cosine_coefficient(0)
f.fourier_series_sine_coefficient(10)
s5 = f.fourier_series_partial_sum(10)
s5.show()
```
