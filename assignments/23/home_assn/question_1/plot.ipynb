```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
n = 4
file = f"/home/r3kste/code/ID2090/assignments/23/home_assn/question_1/data/{n}_obstacles"
```

```python
df = pd.read_csv(f"{file}/all_rrt.csv")
rrts = rrts = ("Normal", "Spaced", "Greedy", "Speedy")
df["Type"] = pd.Categorical(df["Type"], categories=rrts, ordered=True)
df.sort_values(by="Type", inplace=True)
```

```python
sns.catplot(
    data=df,
    x="Path Length",
    y="Type",
    hue="Type",
    kind="violin",
    inner="box",
)
plt.savefig(f"{file}/pathlength.png")
```

```python
sns.scatterplot(
    data=df,
    x="Total Nodes",
    y="Time Taken",
    hue="Type",
    edgecolor=None,
    s=20,
)
plt.savefig(f"{file}/nodesvstime.png")
```

```python
bar_plot = sns.barplot(
    data=df,
    x="Path Length",
    y="Type",
    hue="Type",
    estimator=np.mean,
)
for i in range(len(bar_plot.containers)):
    bar_plot.bar_label(bar_plot.containers[i], label_type="center", fmt="%.2f")
plt.savefig(f"{file}/pathlength_mean.png")
```

```python
sns.kdeplot(
    data=df,
    x="Path Length",
    hue="Type",
    fill=True
)
plt.savefig(f"{file}/pathlength_freq.png")
```
