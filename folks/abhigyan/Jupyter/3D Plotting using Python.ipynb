# Making 3D plots

```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import math
import numpy as np

x=list(map(lambda a: 0.5*a, list(range(25))))
y=x[:]

def gauss(x, x0, y, y0, sd):
    r = (x-x0)**2+(y-y0)**2
    return math.exp(-r/(sd**2))
z = np.zeros([25,25])

for i in range(25):
    for j in range(25):
        z[i,j] = gauss(x[i],3,y[j],3,3) - gauss(x[i],9,y[j],9,3)

fig = plt.figure()
ax = fig.gca(projection='3d')
surf = ax.plot_surface(x,y,z, cmap=cm.hsv, antialiased=True)
ax.view_init(45,90)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
