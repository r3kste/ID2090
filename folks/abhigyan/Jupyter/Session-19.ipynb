# Functions, Differentials and Integrals

```python
var('a,b,c,x')
```

## Derivatives

```python
f(x)=a*x^2+b*x+c
f.show()
```

```python
df(x)=f.derivative(x)
df.show()
```

```python
diff(e^sin(x^2),x).show()
```

```python
g(x)=x^3-x+2
gprime(x)=g(x).derivative()
gprime.show()
plot(g(x),-2,2,color='blue') + plot(gprime(x),-2,2,color='red')
```

## Partial derivatives

```python
f1(x,y) = 3*x^4*y^3 + 9*y*x^2 - 4*x + 8*y
f1.show()
```

```python
df1dx(x,y) = diff(f1,x)
df1dx.show()
```

```python
df1dy(x,y)=diff(f1,y)
df1dy.show()
```

```python
d2f1dxy(x,y)=diff(f1,x,y)
d2f1dxy.show()
```

```python
d2f1dyx(x,y)=diff(f1,y,x)
d2f1dyx.show()
```

## Double differentiation

```python
d2f1dx2(x,y)=diff(f1,x,2)
d2f1dx2.show()
```

## Implicit differentiation

```python
h(x)=function('h')(x)
expr = 5*h^2 + sin(h) == x^2
expr.show()
```

```python
dhdx = solve(diff(expr),diff(h))
```

```python
dhdx[0].show()
```

## Indefinite Integrals

```python
integral(x*sin(x^2),x).show()
```

```python
integral(x/(x^2+1),x).show()
```

```python
integral(1/(1+x^2),x).show()
```

```python
integral(e^(-x^2),x).show()
```

## Integration by partial fractions

```python
expr2 = (x^3-x)/(x^2+5*x+6)
expr2.show()
```

```python
expr2.partial_fraction().show()
```

```python
integral(expr2, x)
```

## Definite Integrals

```python
integral(x/(x^2+1),x,0,1)
```

```python
integral(x^2*e^x, x,0,1)
```

```python
integral(1/x^2,x,2,oo)
```

```python
integral(e^(-x^2),x,-oo,oo)
```

```python
ans = integral((2/sqrt(pi))*exp(-x^2),x,-oo,2)
ans.show()
ans.full_simplify()
```

## Improper Integrals

```python
expr3 = -1 + 1/x^2
expr3.show()
plot(expr3, -10, 10, ymin=-2, ymax=2)
```

```python
integral(expr3,x)
```

```python
integral(1/(1+x^2),x,-oo,oo)
```
