## Laplacian of a scalar field

```python
var('x,y,z')
```

```python
g(x, y, z) = x^2 + y^3 + z^4 + x*y*z^2
```

```python
a=diff(g,2)
```

The output of the above call is Hessian of a scalar field.

```python
print a
```

Laplacian is the trace of Hessian

```python
L = diff(g,2).trace()
```

```python
L.show()
```

## Taylor series for approximation of a function

```python
f(x) = cos(x) + sin(x)
p(x) = taylor(f(x), x, 3*pi/4, 5)
print p(x)
```

```python
p1 = plot( p(x), -3*pi, 3*pi, ymax=5, ymin=-5)
p2 = plot( f(x), -3*pi, 3*pi, linestyle='--')
(p1+p2).show()
```

## Mandelbrot fractal

```python
def m(c):
    z = complex(0,0)
    c = complex(c)
    for j in range(100):
        if abs(z) > 2 :
            break
        z = z^2 + c
    else:
            return complex(0,0)
    return complex(0, j)
```

```python
complex_plot(m, (-2,1), (-1.5, 1.5), plot_points=1000)
```
