# Plotting contours and Quivers

In this lecture, we will learn how to make contour plots and quiver plots.

Now, we will make a list of numbers which will represent our values of x.

```python
x = list(map(lambda a: a*0.5, list(range(0,25,1))))
```

```python
y=x[:]
import math
```

```python
def gauss(x,y,x0,y0,sd):
    r=(x-x0)**2+(y-y0)**2
    return math.exp(-r/(sd**2))
```

Importing Numerical Python to be able to deal with 2 dimensional matrices

```python
import numpy as np
```

Zeros function to generate an empty 2D space with no values

```python
z = np.zeros([25,25])
```

Filling the space with a Gaussian distribution

```python
for i in range(25):
    for j in range(25):
        z[i,j]=gauss(x[i],y[j],3.0,3.0,3.0) - gauss(x[i],y[j],9.0,9.0,3.0)
```

Now, plotting this surface as a contour plot:

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

Plotting with contourf gives us a filled contour. Colorbar gives a legend of the colors used

```python
plt.contourf(x,y,z,10)
plt.colorbar()
plt.xlabel('Distance along x')
plt.ylabel('Distance along y')
plt.title('A Hill and a Valley')
```

Plotting with contour gives us a line contour

```python
plt.contour(x,y,z,10)
plt.colorbar()
plt.xlabel('Distance along x')
plt.ylabel('Distance along y')
plt.title('A Hill and a Valley, but without fill')
```

A clearer hill and valley with more contour intervals

```python
plt.contourf(x,y,z,100)
plt.colorbar()
plt.xlabel('Distance along x')
plt.ylabel('Distance along y')
plt.title('A Hill and a Valley, but clearer')
```

Now, we have plotted a scalar field.

Next we shall plot a vector field, as a quiver plot

```python
gx = np.zeros([24,24])
gy = np.zeros([24,24])
for i in range(1,24):
    for j in range(1,24):
        gx[i,j] = (z[i+1,j]-z[i-1,j])/2
        gy[i,j] = (z[i,j+1]-z[i,j-1])/2
```

Various HSV color maps are feasible. E.g. hot, hsv, rgb, cool, and many more

```python
plt.contourf(z,150,cmap='hs')
plt.colorbar()
plt.quiver(gx,gy)
```
