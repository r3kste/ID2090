# Limits, Series and Transforms

## Limits

```python
var('x')
f(x) = 1/x
plot(f,x,-1,1,ymin=-10, ymax=10)
```

```python
f.limit(x=0,dir='minus')
```

```python
f.limit(x=0,dir='plus')
```

```python
g(x) = (2*x + 8)/(x^2 + x - 12)
g.show()
```

Try evaluating g(-4)

```python
g.limit(x=4)
```

```python
h(x)=(x^2-4)/(x-2)
h.show()
```

```python
h.limit(x=2)
```

```python
ga(x)=x*log(x)+(1-x)*log(1-x)
plot(ga,x,0,1)
```

```python
ga.limit(x=0)
```

```python
ga.limit(x=1)
```

```python
dga(x)=diff(ga,x)
dga.show()
```

```python
dga.limit(x=0)
```

```python
dga.limit(x=1)
```

## Series

```python
s(x)=sin(x)/(x^2)
s.show()
```

Expanding around x=1 using power series

```python
s1(x)=s.series(x==1,4)
s1.show()
```

Expanding using Taylor series

```python
s2(x)=s.taylor(x,1,4)
s2.show()
```

Fourier Series

```python
f = piecewise([((0,pi/2), -1), ((pi/2,pi), 2)])
f.fourier_series_cosine_coefficient(0)
f.fourier_series_sine_coefficient(5)
s5 = f.fourier_series_partial_sum(5);
s5.show()
plot(f, (0,pi)) + plot(s5, (x,0,pi), color='red')
```

## Sum of Series

```python
var('n,k')
```

Alternating harmonic series

```python
ss1(k) = (-1)^(k+1)*1/k
ss1.show()
```

```python
ss1.sum(k,1,infinity)
```

Binomial series

```python
ss2(k)=binomial(n,k)
ss2.show()
```

```python
ss2.sum(k,1,infinity)
```

Harmonic series

```python
ss3 = 1/k
ss3.sum(k,1,infinity)
```

## Transforms

Laplace Transform

```python
var('t, a, k, s')
f(t) = sin(k*t)
F(s) = f.laplace(t,s)
F.show()
```

```python
G(s) = 1/((s-1)*(s+2)*(s+4))
G.show()
g(t) = G.inverse_laplace(s,t)
g.show()
```
