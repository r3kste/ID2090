Here, we implement an algorithm to normalize the difference between random points, and then see the final image

```octave
A = rand(100);
B = zeros(100);

pcolor(A)
```

```octave
for iterator = 1:50
    for i = 2:size(A,1)-1
        for j = 2:size(A,2)-1
            B(i,j) = A(i,j) + 0.2*(A(i+1,j) + A(i-1,j) + A(i,j+1) + A(i,j-1) - 4*A(i,j));
        end;
    end;
    A = B;
    pcolor(A)
end;
```

This code takes a little while to run, but the final output is dark at the edges, and bright at the centre.

Thus, the code is used to reduce the differences of values of points
