# Homework 9 - Q6

## Problem Statement:

> Choose x to be a 1D array with values from 0.01 to 0.99 in intervals of 0.02. Plot a family of
curves of the following function for values of M ranging from 0 to 4 in steps of 0.5.
$$f(x) = x\log(x)+(1-x)\log(1-x)+Mx\log(x)$$

Generating the array `x`:

```octave
x = [0.01:0.02:0.99];
```

`hold on` is used to prevent plotting from occuring in separate graphs.

Thus, we have the following graph for different values of $M$

```octave
hold on
for M = [0:0.5:4]
    y = x.*log(x)+(1.-x).*log(1.-x).+M.*x.*(1.-x);
    plot(x,y)
endfor;
```
