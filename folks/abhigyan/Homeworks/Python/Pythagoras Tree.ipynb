```python
import math
import matplotlib.pyplot as plt
def pythagorasTree(x,y,side,phi,alpha,depth,plots):
    delx = side*math.cos(phi)
    dely = side*math.sin(phi)
    x1 = x + delx
    y1 = y + dely
    x2 = x + delx - dely
    y2 = y + delx + dely
    x3 = x - dely
    y3 = y + delx
    x4 = x3 + side*math.cos(phi+alpha)*math.cos(alpha)
    y4 = y3 + side*math.sin(phi+alpha)*math.cos(alpha)
    xs = [x,x1,x2,x3,x]
    ys = [y,y1,y2,y3,y]
    plots+=plt.plot(xs,ys)
    if depth>1:
        pythagorasTree(x4,y4,side*math.cos(alpha),phi+alpha-math.pi/2,alpha,depth-1,plots)
        pythagorasTree(x3,y3,side*math.cos(alpha),phi+alpha,alpha,depth-1,plots)
plots = plt.plot(0,0)
pythagorasTree(0,0,5,0,math.pi/4,10,plots)
plt.axis("scaled")
```
