```python
t = Tachyon(xres = 1024, yres = 1024, camera_center=(4,0.3,0))
t.light((4,3,2), 0.3, (1,0,1))
t.texture('t0',ambient=0.1,diffuse=0.9,specular=0.5, opacity=1.0,color=(1.0,0,0))
t.texture('t1',ambient=0.1,diffuse=0.9,specular=0.5, opacity=1.0,color=(0,1.0,0))
t.texture('t2',ambient=0.1,diffuse=0.9,specular=0.5, opacity=1.0,color=(0,0,1.0))
t.texture('t3',ambient=0.1,diffuse=0.9,specular=0.5, opacity=1.0,color=(0,1.0,1.0))
k = 0
for i in srange(-1,1.5,0.05):
    k += 1
    t.sphere((i,i^3,i^2),0.1, 't%s'%(k%3))
t.show()
```
