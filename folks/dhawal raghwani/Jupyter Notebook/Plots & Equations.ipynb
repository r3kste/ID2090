## 2-D Plots
We use Gaussian function to visualize

$$ r = \sqrt( (x - x0)^2 + (y-y0)^2) \right) $$
$$ z = \exp\left( -{ r \over \sigma }^2 \right) $$

```python
A = list(map(lambda a: a*0.25, list(range(0,30,1))))
B=A[:]
import math
def gauss(x, y, x0, y0, sd):
    r = (x-x0)**2 + (y-y0)**2;
    return math.exp(-(r/sd)**2)
```

```python
import numpy as np
Z = np.zeros([30,30])
for i in range(0,30,1):
    for j in range(0,30,1):
        Z[i,j] = gauss(A[i], B[j], 5, 5, 2) - gauss(A[i], B[j], 2.5, 2.5, 3)
```

```python
plt.contour(A,B,Z,50)
plt.colorbar()
plt.xlabel('Distance along X')
plt.ylabel('Distance along Y')
plt.title('A Gaussian plot')
```

```python
import numpy as np
import matplotlib.pyplot as plt
gx = np.zeros([24,24]);
gy = np.zeros([24,24]);
for i in range(2,24,1):
    for j in range(2,24,1):
        gx[i,j] = Z[i+1,j] - Z[i-1,j];
        gy[i,j] = Z[i,j+1] - Z[i,j-1];
```

```python
plt.contourf(Z,50,cmap='hsv')
plt.colorbar()
plt.quiver(gx,gy)
```
