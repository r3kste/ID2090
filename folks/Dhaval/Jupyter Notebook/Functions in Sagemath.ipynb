```python
@interact
def myplot(f=input_box(default=x^2), auto_update=False):
    p = plot(f,(x,-5,5))
    p.show()
```

```python
@interact
def myfact(n=slider(1,20, step_size=0.5), auto_update=False):
    p = plot(sin((1.5+n)*x),(x,0,2*pi), axes_labels=["x", "sin((1.5+n)*x)"])
    p.show()
```

```python
@interact
def mytaylor(
    f = input_box(default=e^x),
    n = slider(vmin=0,vmax=20,step_size=1,default=3,label="select order:"),
    x0 = input_box(default=0),
    simplified = selector(values=["yes","No"], label="Simplify?", default="Yes"),
    auto_update=False
    
):
    
    if(simplified == "Yes"):
        fig1a = plot(f,(x,0,2*pi),color='red')
        fig1b = plot(f.taylor(x,x0,n),(x,0,2*pi),color='blue')
        (fig1a + fig1b).show()
        print(f, " = ", f.taylor(x,x0,n).full_simplify())
    else:
        print(f," = ", f.taylor(x,x0,n))
```
